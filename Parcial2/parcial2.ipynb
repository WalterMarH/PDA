```python
def getCuadrantes(archivo):
    filas=archivo.shape[0]
    columnas=archivo.shape[1]
    cuad1=archivo[0:int(filas/2),0:int(columnas/2)]
    cuad2=archivo[int(filas/2):filas,0:int(columnas/2)]
    cuad3=archivo[0:int(filas/2),int(columnas/2):columnas]
    cuad4=archivo[int(filas/2):filas,int(columnas/2):columnas] 
    return cuad1, cuad2, cuad3, cuad4
```

```python
cuadrantes=getCuadrantes(data3)
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2,figsize=(8,8))
ax[0,0].imshow(cuadrantes[0])
ax[1,0].imshow(cuadrantes[1])
ax[0,1].imshow(cuadrantes[2])
ax[1,1].imshow(cuadrantes[3])
plt.show()
```
