## <span style='color:#55aa74'>1. Analizando datos de eficiencia energética (obligatorio)</span>

Cargue el archivo *eficienciaEnergética.csv*. Agregue una columna llamada *superficieTotal* que sea la suma de las columnas *Surface_Area, Wall_Area, Roof_Area*. Agreuge una columna llamada *cargaTotal* que sea la suma de *Heating_Load, Cooling_Load*.

Luego,

- Realice un gráfico de dispersión comparando *superficieTotal* con *cargaTotal*. Utilice los datos de la columna *Orientation* para diferenciar cada punto de la gráfica según la orientación de la casa.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
````

````python
## TODO 1
efic = pd.read_csv('datasets/eficienciaEnergética.csv')


````

````python
superficieTotal=efic["Wall_Area"]+efic["Roof_Area"]+efic["Surface_Area"]
cargaTotal=efic["Heating_Load"]+efic["Cooling_Load"]
efic["superficieTotal"]=superficieTotal
efic["cargaTotal"]=cargaTotal
efic
````

````python
fig, ax = plt.subplots(figsize=(10,7))
plt.xlabel("Área Total")
plt.ylabel("Carga Total")
colores=["black","black","red","green","blue","yellow"]
label=["","","Norte","Sur","Este","Oeste"]
for i in range(2,6):#efic["Orientation"].unique():
    superficie=efic[efic["Orientation"]==i]["superficieTotal"]
    carga=efic[efic["Orientation"]==i]["cargaTotal"]
    ax.scatter(superficie,carga,c=colores[i],label=label[i])
ax.legend()
````

## <span style='color:#55aa74'>2. Personas en situación de calle (obligatorio)</span>

Utilizando los datos dentro de *sinhogares.csv* se le pide.

- Un gráfico de barras verticales donde se muestre el valor medio de individuos en situación de calles separados por *región*. Agregue el desvío estándar a cada barra.

````python
## TODO 2
sinhogares = pd.read_csv('datasets/sinhogares.csv')
sinhogares
````

````python
indices=[]
medias=[]
desviaciones=[]
for i in sinhogares["region"].unique():
    media=sinhogares[sinhogares["region"]==i]["individuos"].mean()
    desv=sinhogares[sinhogares["region"]==i]["individuos"].std()
    indices.append(i)
    medias.append(media)
    desviaciones.append(desv)

plt.bar(indices,medias,yerr=desviaciones)
plt.tight_layout(rect=(0,0,3,2))
plt.xlabel("Regiones")
plt.ylabel("Individuos en situación de calle")
````

````python
ligas = ['Spain Primera Division', 'Italian Serie A', 'German 1. Bundesliga', 'French Ligue 1', 'English Premier League', 'Holland Eredivisie', 'USA Major League Soccer', 'Chinese Super League', 'Saudi Abdul L. Jameel League', 'Turkish Süper Lig', 'Argentina Primera División', 'Uruguayan Primera División']
fig, ax = plt.subplots(figsize=(35,8))
costos=[]
for i in ligas:
    costo=fifa21[fifa21['league_name']==i]["value_eur"].sum()
    costos.append(costo)
ax.bar(ligas,costos)
````

````python
## TODO 5.5.1
pociciones=fifa21["player_positions"].values
nuevaLista=[]
cuenta=0
for i in pociciones:
    i=(i.translate({ord(' '): None}))
    i=str(i).split(",")
    if i[0]=="ST" or i[0]=="RW" or i[0]=="LW" or i[0]=="CF" or i[0]=="RF" or i[0]=="LF":
        nuevaLista.append("Atacante")
        cuenta+=1
    elif i[0]=="CAM" or i[0]=="CM" or i[0]=="RM" or i[0]=="LM"or i[0]=="CDM":
        nuevaLista.append("Mediocampista")
        cuenta+=1
    elif i[0]=="RWB" or i[0]=="RB" or i[0]=="CB" or i[0]=="LB" or i[0]=="LWB":
        nuevaLista.append("Defensa")
        cuenta+=1
    elif i[0]=="GK":
        cuenta+=1
        nuevaLista.append("Arquero")
    else:
        print(i)
fifa21["onePosition"]=nuevaLista
fifa21.drop("player_positions",inplace=True,axis=1)
````

````python
#- Repita el ejercicio 5.1 de la *guía de ejercicios de Pandas*. Luego realice un gráfico 
# de dispersión en donde se compare el precio de cada jugador con su *overall*. Discrimine los colores de los puntos en base a
#  la columna *onePosition* obtenida luego de resolver la parte 5.1.
fig, ax = plt.subplots(figsize=(18,8))
posiciones=fifa21["onePosition"].unique()
for i in posiciones:
    overall=fifa21[fifa21["onePosition"]==i]["overall"]
    precio=fifa21[fifa21["onePosition"]==i]["value_eur"]
    ax.scatter(overall,precio,label=i,alpha=0.5)
ax.set_facecolor("darkslategray")
ax.legend()

    
````
